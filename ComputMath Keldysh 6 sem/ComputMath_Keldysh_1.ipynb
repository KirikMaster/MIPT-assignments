## 7 задача

```python
A = np.array([[2, 3], [1, 4]])
print(A)
print(np.linalg.det(A))
print(np.matrix.trace(A))
B = A@A@A.T - A@A.T
print("determinant B: ", np.linalg.det(B))
print("trace B: ", np.matrix.trace(B))
eigvalA, eigvalB = np.linalg.eigvals(A), np.linalg.eigvals(B)
print("Eigenvalues A, B: ", eigvalA, eigvalB)
C = A + np.eye(2)
print("C: ", C)
D = B + np.ones((2,2))
print("D: ", D)
M = C @ D.T + A @ A - np.linalg.inv(A) @ B @ B @ B
print("M, det(M), trace(M) : ", '\n', M, '\n', np.linalg.det(M), np.matrix.trace(M))
print("Eigenvalues M: ", np.linalg.eigvals(M))
v = M[:,1]
print("v = ", v)
print("sum(v), multiplication(v)", np.sum(v), np.prod(v))
```

## 23 Задача

```python
p = 1.0
e = [0, 0.1, 0.2, 0.4, 0.6]

def rad(theta, n):
    return p / (1 + e[n] * np.cos(theta))

def coords(r, theta):
    return (r*np.cos(theta), r*np.sin(theta))
```

```python
theta = np.linspace(0, 2*np.pi, 100)
fig, ax = plt.subplots()
for i in range(len(e)):
    X, Y = coords(rad(theta, i), theta)
    ax.plot(X, Y, label=('e = ' + str(e[i])), linewidth=2.0)
ax.set_title('Орбита в полярных координатах', size=12)
ax.grid()
ax.legend()
ax.set_xlabel('Угол', size=12)
ax.set_ylabel('Радиус', size=12)
plt.show
fig.savefig('polar.png')
```
