```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
```

```python
digits = load_digits(as_frame=True)
digits.data
```

### 1 Задание
Метод PCA:

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(digits.data)
df = pd.DataFrame(pca_result, columns=['pca-one', 'pca-two', 'pca-three'])
df['target'] = digits.target
df
```

```python
df.describe()
```

### 2 Задание
Метод t-SNE:

```python
tsne = TSNE(n_components=2)
tsne_results = tsne.fit_transform(digits.data)
df_tsne = pd.DataFrame(tsne_results, columns=['tsne-2d-one', 'tsne-2d-two'])
df_tsne['target'] = digits.target
df_tsne
```

```python
fig = plt.figure(figsize=(20, 8))

rect = 0, 0, 0.5, 1
ax_pca = fig.add_axes(rect, projection='3d')
ax_pca.scatter(df['pca-one'], df['pca-two'], df['pca-three'], c=df['target'], cmap='viridis')
ax_pca.set_title('PCA Visualization', size=20)

rect = 0.5, 0, 0.5, 1
ax_tsne = fig.add_axes(rect)
ax_tsne.scatter(df_tsne['tsne-2d-one'], df_tsne['tsne-2d-two'], c=df_tsne['target'], cmap='viridis')
ax_tsne.set_title('t-SNE Visualization', size=20)

fig.colorbar(ax=ax_tsne, mappable=None, boundaries=np.arange(0, 11, 1), values=np.arange(0, 10, 1))

plt.show()
```

```python
df_per = []
perplexities = (1, 3, 10, 30)
for perplexity in perplexities:
    tsne = TSNE(n_components=2, perplexity=perplexity)
    tsne_results = tsne.fit_transform(digits.data)
    df_per.append(pd.DataFrame(tsne_results, columns=['tsne-2d-one', 'tsne-2d-two']))
    df_per[-1]['target'] = digits.target
```

```python
fig2, ax2 = plt.subplots(2, 2, figsize=(12, 10))
df_iter = iter(df_per)
perplexities_iter = iter(perplexities)
for ax in ax2.reshape(4):
    dfi = next(df_iter)
    ax.scatter(dfi['tsne-2d-one'], dfi['tsne-2d-two'], c=dfi['target'], cmap='viridis')
    ax.set_title(f'Perplexity = {next(perplexities_iter)}', size=12)
    fig.colorbar(ax=ax, mappable=None, boundaries=np.arange(0, 11, 1), values=np.arange(0, 10, 1))
plt.plot()
```
