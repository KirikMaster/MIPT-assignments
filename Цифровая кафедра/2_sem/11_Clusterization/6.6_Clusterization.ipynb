```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, FeatureAgglomeration, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

### 1 Задание
Метод K-Means:

```python
df = pd.read_csv('dataset1.csv')
plt.scatter(df['F0'], df['F1'], c=df['LABEL'])
plt.grid()
plt.colorbar()
df
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(df[['F0', 'F1']], y=df['LABEL'])
kmeans.cluster_centers_
```

```python
_, uc = np.unique(clusters, return_counts=True)
print(uc)
```
